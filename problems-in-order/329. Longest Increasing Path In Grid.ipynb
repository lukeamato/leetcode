# [329. Longest Increasing Path in a Matrix](https://leetcode.com/problems/longest-increasing-path-in-a-matrix/description/?envType=company&envId=doordash&favoriteSlug=doordash-six-months)

Given an `m x n` integers `matrix`, return the length of the longest increasing path in `matrix`.

From each cell, you can either move in four directions: left, right, up, or down. You **may not**  move **diagonally**  or move **outside the boundary**  (i.e., wrap-around is not allowed).

**Example 1:** 
<img alt="" src="https://assets.leetcode.com/uploads/2021/01/05/grid1.jpg" style="width: 242px; height: 242px;">

```
Input: matrix = [[9,9,4],[6,6,8],[2,1,1]]
Output: 4
Explanation: The longest increasing path is `[1, 2, 6, 9]`.
```

**Example 2:** 
<img alt="" src="https://assets.leetcode.com/uploads/2021/01/27/tmp-grid.jpg" style="width: 253px; height: 253px;">

```
Input: matrix = [[3,4,5],[3,2,6],[2,2,1]]
Output: 4
Explanation: The longest increasing path is `[3, 4, 5, 6]`. Moving diagonally is not allowed.
```

**Example 3:** 

```
Input: matrix = [[1]]
Output: 1
```

**Constraints:** 

- `m == matrix.length`
- `n == matrix[i].length`
- `1 <= m, n <= 200`
- `0 <= matrix[i][j] <= 2^31 - 1`

````python
# class Solution:
#     def longestIncreasingPath(self, matrix: List[List[int]]) -> int:
#         @cache
#         def increasingPath(start: tuple) -> int:
#             directions = [(-1,0),(1,0),(0,1),(0,-1)]

#             x, y = start
#             longestPath = 1
#             for dx, dy in directions:
#                 if (0 <= x+dx < len(matrix) 
#                     and 0 <= y+dy < len(matrix[0]) 
#                     and matrix[x][y] < matrix[x+dx][y+dy]):
#                     longestPath = max(longestPath, 1+increasingPath((x+dx,y+dy)))
#             return longestPath


#         longestPath = 1
#         for i in range(len(matrix)):
#             for j in range(len(matrix[0])):
#                 longestPath = max(longestPath, increasingPath((i,j)))
#         return longestPath

class Solution:
    def longestIncreasingPath(self, matrix: list[list[int]]) -> int:
        memo = [[None for _ in range(len(matrix[0]))] for _ in range(len(matrix))]

        def increasingPath(start: tuple) -> int:
            directions = [(-1,0),(1,0),(0,1),(0,-1)]

            x, y = start
            longestPath = 1
            for dx, dy in directions:
                if (0 <= x+dx < len(matrix) 
                    and 0 <= y+dy < len(matrix[0]) 
                    and matrix[x][y] < matrix[x+dx][y+dy]):
                    if memo[x+dx][y+dy]:
                        longestPath = max(longestPath, 1+memo[x+dx][y+dy])
                    else:
                        longestPath = max(longestPath, 1+increasingPath((x+dx,y+dy)))
            memo[x][y] = longestPath
            return longestPath


        longestPath = 1
        for i in range(len(matrix)):
            for j in range(len(matrix[0])):
                longestPath = max(longestPath, increasingPath((i,j)))
        return longestPath
````

### Time complexity
 O(V) = O(m * n) as we are doing DFS with a matrix that only has 4 edges per node. Same as other similar problems

### Space complexity 
Also only O(m * n) because the memo is O(m * n) space and the call stack for recursive DFS will be O(m * n) at most"
